```python
# tehdään dataframe newdf jossa on omilla riveillään quoted ja replied_to yhteydet dataframesta q_rt

newdf = pd.DataFrame(columns=["Source", "Target", "referenced_tweets.type"])

for i in range(len(q_rt)):
    source = q_rt.loc[i]["Source"]
    target1 = int(q_rt.loc[i]["Target"].split(";")[0])
    target2 = int(q_rt.loc[i]["Target"].split(";")[1])
    rows = pd.DataFrame([[source, target1, "quoted"], [source, target2, "replied_to"]], columns=["Source", "Target", "referenced_tweets.type"])
    newdf = pd.concat([newdf, rows], ignore_index=True)
    
```
